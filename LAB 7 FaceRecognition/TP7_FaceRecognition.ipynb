````
hog_detector = dlib.get_frontal_face_detector()
cnn_detector = dlib.cnn_face_detection_model_v1('models/mmod_human_face_detector.dat')
#   About ==== dlib.get_frontal_face_detector() ====
#   This example program shows how to find frontal human faces in an image.  In
#   particular, it shows how you can take a list of images from the command
#   line and display each on the screen with red boxes overlaid on each human
#   face.

def face_locations(image, model="hog"):
    """
    Retourne une liste contenant pour chaque images un cadre autour du visage
    @param: image est un tableau np
    """
    
    if model == "hog":
        detector = hog_detector
        cst = 0
    elif model == "cnn":
        detector = cnn_detector
        cst = 10
            
    matches = detector(image,1)
    rects   = []
    
    for r in matches:
        if model == "cnn":
            r = r.rect
        x = max(r.left(), 0)
        y = max(r.top(), 0)
        w = min(r.right(), image.shape[1]) - x + cst
        h = min(r.bottom(), image.shape[0]) - y + cst
        rects.append((x,y,w,h))
        
    return rects
````

````
def extract_faces(image, model="hog") -> np.array([]):
    """
    Retourne l'image du visage zoom√© et centr√©
    @param: image est un tableau np
    @return: faces : List de tableau Numpy de taille 128*128
    """
    
    gray  = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    rects = face_locations(gray, model)
    faces = []
    
    for (x,y,w,h) in rects:
        cropped = image[y:y+h, x:x+w, :]
        cropped = cv2.resize(cropped, (128,128))
        faces.append(cropped)
            
    return faces
````

````
def list_images(basePath, validExts=(".jpg", ".jpeg", ".png", ".bmp", ".tif", ".tiff"), contains=None):
    """
    Retourne une liste de path des images, de tous les rep et ss rep.
    """
    
    imagePaths = []
    
    # loop over the directory structure
    for (rootDir, dirNames, filenames) in os.walk(basePath):
        # loop over the filenames in the current directory
        for filename in filenames:
            # if the contains string is not none and the filename does not contain
            # the supplied string, then ignore the file
            if contains is not None and filename.find(contains) == -1:
                continue

            # determine the file extension of the current file
            ext = filename[filename.rfind("."):].lower()

            # check to see if the file is an image and should be processed
            if ext.endswith(validExts):
                # construct the path to the image and yield it
                imagePath = os.path.join(rootDir, filename).replace(" ", "\\ ")
                imagePaths.append(imagePath)
    
    return imagePaths
````

````
def labels_to_cat(labels) -> np.ndarray :
    """
    Create a hot one encoder for each label.
    Format : vector of len = nb categories
    :param labels: <!> take string array as input
    :return: a vector like this: [1, 0, 0, 0]. It mean the category is 1
    """
    ''' ==== Documentation ====
    1. https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.LabelEncoder.html
    '''
    # Create a LabelEncoder object and fit it to the labels
    le = LabelEncoder().fit(labels)

    # transform the string labels to integers using the LabelEncoder
    integer_labels = le.transform(labels)

    # Convert the integer labels to categorical labels using the to_categorical function
    categorical_labels = to_categorical(integer_labels)

    return categorical_labels
````

````
def split_data(values: np.ndarray, labels: np.ndarray, test_size: float = 0.3, random_state: int = RANDOM_STATE): 
    """
    split randomly the given data into 4 set.
    :param values: values
    :param labels: labels
    :param test_size: the size of each set => test = 0.3 by default
    :param random_state: To keep an order in the random split. Is useful to compare each runs
    :return: faces_train, faces_test, labels_train, labels_test
    """
    ''' ==== Documentation ====
    1. https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html
    '''

    faces_train, faces_test, labels_train, labels_test = train_test_split(
        values,
        labels,
        test_size=test_size,
        random_state=random_state
        )

    return faces_train, faces_test, labels_train, labels_test
````

````
#@title ##### Batch preparation
def make_dataset(data, batch_size : int = BATCH_SIZE):
    """
    BUILD THE DATASET : * Use data augmentation
    1. Data augmentation => ImageDataGenerator
    2. Batches preparation => flow
    This is done for the train and the test
    :param data: the split data in train/test and values/labels
    :param batch_size: the size of the batches
    :return: train_dataset, test_dataset
    """

    ''' ==== Documentation ====
    1. https://www.tensorflow.org/api_docs/python/tf/keras/preprocessing/image/ImageDataGenerator
    2. https://studymachinelearning.com/keras-imagedatagenerator-with-flow/
    '''

    faces_train, faces_test, labels_train, labels_test = data

    # 1. Data augmentation
    train_data_augmentation = image.ImageDataGenerator(
        rescale=1. / 255,
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest'
    )

    test_data_augmentation = image.ImageDataGenerator(rescale=1. / 255)


    # 2. Create the batches for the training and testing sets
    train_dataset = train_data_augmentation.flow(
        faces_train,
        labels_train,
        batch_size=batch_size,
        ignore_class_split=True,
        subset='training'
    )


    test_dataset = test_data_augmentation.flow(     # si prb changer test_data_augmentation par train_data_augmentation <!>
        faces_test,
        labels_test,
        batch_size=batch_size,
        ignore_class_split=True,
        # subset='validation'               # TODO : la size de validation est 0 si on precise l'argument subset ?!
    )


    return train_dataset, test_dataset
````

````
#@title #### Plot function
def history_plot(history, title : str = "", plt_name : str = "", save : bool = False):
    """
    Plot two sub-graph to visualize the accuracy and the loss for a cnn.
    Display title and subtitle + all metrics
    :param history: a history object
    :param title: The title of the plot
    :param plt_name: the name of the plot
    :param save: if is true the plot will be saved on Google Drive
    :return:
    """
    # Get the values
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']

    # make the pair plot
    fig, axs = plt.subplots(1, 2, figsize=(30,5))
    fig.suptitle(title, fontsize=14, fontweight='bold')
    fig.subplots_adjust(top=0.85)

    # values
    axs[0].set_title("Accuracy", fontsize=12)
    axs[0].plot(acc, color='b', label='Training accuracy')
    axs[0].plot(val_acc, color='m', label='Validation accuracy')
    axs[0].set_xlabel('Epochs')
    axs[0].set_ylabel('Accuracy')
    axs[0].legend()

    axs[1].set_title("Loss", fontsize=12)
    axs[1].plot(loss, color='b', label='Training loss')
    axs[1].plot(val_loss, color='m', label='Validation loss')
    axs[1].set_xlabel('Epochs')
    axs[1].set_ylabel('Loss')
    axs[1].legend()


    # description : get metrics 
    ACC         = round(history.history['accuracy'][-1], 2)
    VAL_ACC     = round(history.history['val_accuracy'][-1], 2)
    LOSS        = round(history.history['loss'][-1], 2)
    VAL_LOSS    = round(history.history['val_loss'][-1], 2)

    # add information
    description = f"\nACC : {ACC}\nVAL_ACC : {VAL_ACC}\nLOSS : {LOSS}\nVAL_LOSS : {VAL_LOSS}"
    fig.text(0.15, -0.2, description, ha='left')
    
    # save the plt to the drive
    if save : 
        try : 
            if not os.path.exists(drive_plots):
                os.makedirs(drive_plots)

            path = os.path.join(drive_plots, plt_name)
            plt.savefig(path, dpi=300, bbox_inches='tight')

        except Exception as e:
            print(e)
            plt.close(fig)


    plt.show()
    return
````

````
#@title ##### CNN
def CNN(train_generator, test_generator, version_number : int, epochs=EPOCHS, save : bool = SAVE_NETWORK) : 
    """
    Train and evaluate a convolutional neural network on the train and test ImageDataGenerator objects.
    **It is a complet CNN to preform classification**
    :param train_generator: The train ImageDataGenerator object
    :param test_generator: The test ImageDataGenerator object
    :param version_number: The version of the architecture
    :param epochs: The number of epochs the network will performe
    :param save: if is true the plot will be saved on Google Drive
    :return: history object corresponding to the CNN
    """

    # Define the architecture of the convolutional neural network
    model = models.Sequential()
    model.add(layers.Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(128, 128, 3)))
    model.add(layers.MaxPooling2D(pool_size=(2, 2)))
    model.add(layers.Conv2D(64, kernel_size=(3, 3), activation='relu'))
    model.add(layers.MaxPooling2D(pool_size=(2, 2)))
    model.add(layers.Conv2D(128, kernel_size=(3, 3), activation='relu'))
    model.add(layers.MaxPooling2D(pool_size=(2, 2)))
    model.add(layers.Flatten())
    model.add(layers.Dropout(0.5))
    model.add(layers.Dense(512, activation='relu'))
    model.add(layers.Dense(6, activation='softmax')) 

    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

    history = model.fit(
      train_dataset, 
      epochs=epochs, 
      validation_data=test_dataset)
    
    if save : 
        model.save(os.path.join(drive_models, f"CNN_{version_number}.h5"))
        store_on_drive(history, f"CNN_{version_number}", drive_history)

    return history
````

## 2. Pose estimation

You have isolated the faces in our image. But now you have to deal with the problem that faces turned different directions look totally different to a computer. To account for this, you will try to warp each picture so that the eyes and lips are always in the same place in the image. More concretely, you are going to use an algorithm called face landmark estimation. The basic idea is to locate 68 specific points (called landmarks) that exist on every face: ‚Ääthe top of the chin, the outside edge of each eye, the inner edge of each eyebrow, etc. Then, you‚Äôll simply rotate, scale and shear the image so that the eyes and mouth are centered as best as possible. This will make face recognition more accurate.

````
pose68 = dlib.shape_predictor('models/shape_predictor_68_face_landmarks.dat')
pose05 = dlib.shape_predictor('models/shape_predictor_5_face_landmarks.dat')

def face_landmarks(face, model="large"):
    
    if model == "large":
        predictor = pose68
    elif model == "small":
        predictor = pose05
    
    if not isinstance(face, list):
        rect = dlib.rectangle(0,0,face.shape[1],face.shape[0])
        return predictor(face, rect)
    else:
        rect = dlib.rectangle(0,0,face[0].shape[1],face[0].shape[0])
        return [predictor(f,rect) for f in face]
````

## 3. Face encoding

The simplest approach to face recognition is to directly classify an unknown face with a convnet trained on your database of tagged people. Seems like a pretty good idea, right? There‚Äôs actually a huge problem with that approach. A site like Facebook with billions of users and a trillion photos can‚Äôt possibly train such a big convnet. That would take way too long. What you need is a way to extract a few basic measurements from each face, which you can then use to quickly compare the unknown face with your database. For example, you might measure the size of each ear, the spacing between the eyes, the length of the nose, etc. However, it turns out that the measurements that seem obvious to us humans (like eye color) don‚Äôt really make sense to a computer looking at individual pixels in an image. Researchers have discovered that the most accurate approach is to let the computer figure out the measurements to collect itself. Deep learning does a better job than humans at figuring out which parts of a face are important to measure.

The solution is to train a convnet. But instead of training the network to classify pictures, it is trained to generate 128 measurements for each face. The training process works by looking at 3 face images at a time: the picture of a known person, another picture of the same known person, and a picture of a totally different person. Then, the algorithm looks at the measurements currently generated for each of those three images. It tweaks the neural network slightly to make sure that the measurements generated for the same person are slightly closer, and the measurements for different persons are slightly further apart. After repeating this step millions of times for millions of images of thousands of different people, the convnet learns to generate 128 measurements for each person. 

<img src="https://perso.esiee.fr/~najmanl/FaceRecognition/figures/triplet.png" style="height:400px;">

This process of training a convnet to output face encodings requires a lot of data and computer power. Even with an expensive GPU, it takes about 24 hours of continuous training to get good accuracy. But once the network has been trained, it can generate measurements for any face, even ones it has never seen before! So this step only needs to be done once. Fortunately, the people at [OpenFace](https://cmusatyalab.github.io/openface/) already did this and they published several trained networks which you can directly use. So all you need to do is run your face images through their pre-trained network to get the 128 measurements for each face.

<img src="https://perso.esiee.fr/~najmanl/FaceRecognition/figures/encoding.png" style="height:300px;">

````
#@title ###### KNN classifieur
def classifier_KNN(values_train, labels_train, values_to_pred : np.ndarray, print_score: bool = True, n_neighbors: int = 7) -> tuple:
    """
    classifier_KNN predicts labels after training on a dataset: values_train, labels_train.
    :param values_train: .ndim == 2 | nd.array of images
    :param labels_train: Vector of int (unique int per class)
    :param values_to_pred: .ndim == 2 | nd.array of images
    :param print_score: display the labels for each image, the execution time, the accuracy and the confusion matrix
    :param n_neighbors: The number of neighbors
    :return: labels_pred, classifier_time
    """

    if values_to_pred.ndim == 1:
        values_to_pred = np.array([values_to_pred])
    assert values_train.ndim == 2 and labels_train.ndim == 1 and values_to_pred.ndim == 2, "The dimensions are wrong !"

    # *******   classifier computation   *******
    cla_knn = make_pipeline(StandardScaler(), KNeighborsClassifier(n_neighbors))
    cla_knn.fit(values_train, labels_train)
    score_knn = cla_knn.score(values_train, labels_train)


    start_time = time.time()
    labels_pred = cla_knn.predict(values_to_pred)
    classifier_time = time.time() - start_time

    # *******   print metrics   *******
    print(f"\nclassifier_KNN : ({n_neighbors})")
    if print_score:
        print("Images number : {}".format([i+1 for i in range(values_to_pred.shape[0])]))
        print("Images classe : {}".format([i for i in labels_pred]))
    print("Execution time : {:.3f} ms".format(classifier_time*100))
    print("Score  \t       : {:.2f} %".format(score_knn*100))
    print("-----------------------------------")

    return labels_pred, classifier_time
````

````
#@title ###### SVM classifieur
def classifier_SVM(values_train, labels_train, values_to_pred, print_score=True, kernel_type : str = 'linear') -> tuple:
    """
    classifier_SVM predicts labels after training on a dataset: values_train, labels_train.
    :param values_train: .ndim == 2 | nd.array of images
    :param labels_train: Vector of int (unique int per class)
    :param values_to_pred: .ndim == 2 | nd.array of images
    :param print_score: display the labels for each image, the execution time, the accuracy and the confusion matrix
    :param kernel_type: Precise the kernel type : 'linear', 'rbf', and so on...
    :return: labels_pred, classifier_time
    """

    assert values_train.ndim == 2 and labels_train.ndim == 1 and values_to_pred.ndim == 2, "The dimensions are wrong !"

    # *******   classifier computation   *******
    cla_svm = make_pipeline(StandardScaler(), SVC(kernel=kernel_type))
    cla_svm.fit(values_train, labels_train)
    score_svm = cla_svm.score(values_train, labels_train)

    start_time = time.time()
    labels_pred = cla_svm.predict(values_to_pred)
    classifier_time = time.time() - start_time

    # *******   print metrics   *******
    print(f"\nclassifier_SVM : ('{kernel_type}')")
    if print_score:
        print("Images number : {}".format([i+1 for i in range(values_to_pred.shape[0])]))
        print("Images classe : {}".format([i for i in labels_pred]))
    print("Execution time : {:.3f} ms".format(classifier_time*100))
    print("Score  \t       : {:.2f} %".format(score_svm*100))
    print("-----------------------------------")


    return labels_pred, classifier_time
````

````
#@title #### detect a face in a frame and label it with a classifier
def process_frame(image, mode="fast"):
    
    # face detection
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    if mode == "fast":
        matches = hog_detector(gray,1)
    else:
        matches = cnn_detector(gray,1)
        matches = [m.rect for m in matches]
        
    for rect in matches:
        
        # face landmarks
        landmarks = pose68(gray, rect)
        
        # face encoding
        encoding = cnn_encoder.compute_face_descriptor(image, landmarks)
        
        # face classification
        label = str(classifier.predict([encoding]))
        
        # draw box
        cv2.rectangle(image, (rect.left(), rect.top()), (rect.right(), rect.bottom()), (0, 255, 0), 2)
        y = rect.top() - 15 if rect.top() - 15 > 15 else rect.bottom() + 25
        cv2.putText(image, label, (rect.left(), y), cv2.FONT_HERSHEY_SIMPLEX, 0.75, (0, 255, 0), 2)
        
    return image
````

````
#@title ##### for movies
def process_movie(video_name, mode="fast"):
    
    video  = cv2.VideoCapture(video_name)
    
    try:
        
        while True:

            # Grab a single frame of video
            ret, frame = video.read()
            
            # Resize frame of video for faster processing
            frame = cv2.resize(frame, (0, 0), fx=0.5, fy=0.5)

            # Quit when the input video file ends or key "Q" is pressed
            key = cv2.waitKey(1) & 0xFF
            if not ret or key == ord("q"):
                break

            # Process frame
            image = process_frame(frame, mode)

            # Display the resulting image
            cv2_imshow(image)
    
    finally:
        video.release()
        cv2.destroyAllWindows()
        print("Video released")
````

````
#@title ##### take a picture with the webcam of the laptop
def take_photo(path_to_directory = custom_dataset, firstname : str = 'John', lastname : str = 'Doe', quality=0.8):

  path_to_image = path_setter(path_to_directory, firstname, lastname)


  js = Javascript('''
    async function takePhoto(quality) {
      const div = document.createElement('div');
      const capture = document.createElement('button');
      capture.textContent = 'Capture';
      div.appendChild(capture);

      const video = document.createElement('video');
      video.style.display = 'block';
      const stream = await navigator.mediaDevices.getUserMedia({video: true});

      document.body.appendChild(div);
      div.appendChild(video);
      video.srcObject = stream;
      await video.play();

      // Resize the output to fit the video element.
      google.colab.output.setIframeHeight(document.documentElement.scrollHeight, true);

      // Wait for Capture to be clicked.
      await new Promise((resolve) => capture.onclick = resolve);

      const canvas = document.createElement('canvas');
      canvas.width = video.videoWidth;
      canvas.height = video.videoHeight;
      canvas.getContext('2d').drawImage(video, 0, 0);
      stream.getVideoTracks()[0].stop();
      div.remove();
      return canvas.toDataURL('image/jpeg', quality);
    }
    ''')
  display(js)
  data = eval_js('takePhoto({})'.format(quality))
  binary = b64decode(data.split(',')[1])
  with open(path_to_image, 'wb') as f:
    f.write(binary)
  return path_to_image
````

````
#@title ##### Make the folder and return the path to the image
def path_setter(path_to_directory = custom_dataset, firstname : str = 'John', lastname : str = 'Doe'): 
    """
    Cr√©e le path pour chaque image, avec un num√©ro unique. Si le dossier n'existe pas il est cr√©√©e.
    :param path_to_directory: path to corresponding directory
    :param firstname: The first name of the personne
    :param lastname: The lase name of the personne
    :return: path_to_file
    """
    increment = 0
    firstname = firstname.casefold()
    firstname = firstname.replace(" ", "")
    lastname = lastname.casefold()
    lastname = lastname.replace(" ", "")
    people_name = firstname + lastname

    assert people_name.isalpha(), "Le nom ne contient pas que des lettres ! "

    path_to_ppl = os.path.join(path_to_directory, people_name)
    # Create the directory if it does not exist
    if not os.path.exists(path_to_ppl):
        os.makedirs(path_to_ppl)

    if len(os.listdir(path_to_ppl)) != 0:
        increment = int(os.listdir(path_to_ppl)[-1].split('_')[-1].split('.')[0]) + 1
                    # split on the  '_' et then on  '.' ‚Üí recuperate le number

        assert isinstance(increment, int), "le num√©ro de l'image est introuvable"
    
    path_to_file = os.path.join(path_to_ppl, '{}_{}{}'.format(people_name, increment, '.jpeg'))

    return path_to_file
````
